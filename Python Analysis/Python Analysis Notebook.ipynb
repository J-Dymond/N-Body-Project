```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

#To increase figure size -> when putting into report
#plt.rcParams['figure.figsize'] = [10, 10]

import numpy as np
import pylab 

#Plotting position and oribtal distance
OrbitData = open('../Data/DataStable.txt', 'r') 
Positions = []
for line in OrbitData.readlines():
    values = line.split()
    for i in range(len(values)):
        values[i] = float(values[i])
    Positions.append(values)
OrbitData.close()
Positions = np.array(Positions)

time = np.array(Positions[:,0])

SunPos = np.array(Positions[:,1:3])
EarthPos = np.array(Positions[:,3:5]) - SunPos
MarsPos = np.array(Positions[:,5:7]) - SunPos
JupiterPos = np.array(Positions[:,7:9]) - SunPos
SaturnPos = np.array(Positions[:,9:11]) - SunPos
UranusPos = np.array(Positions[:,11:13]) - SunPos
NeptunePos = np.array(Positions[:,13:15]) - SunPos
PerturberPos = np.array(Positions[:,15:17]) - SunPos
PerturberApproach = np.array(Positions[:,17])
print(PerturberApproach.min())

plt.plot(EarthPos[:,0], EarthPos[:,1], '-',label='Earth')
plt.plot(MarsPos[:,0], MarsPos[:,1], '-',label='Mars')
plt.plot(JupiterPos[:,0], JupiterPos[:,1], '-',label='Jupiter')
plt.plot(SaturnPos[:,0], SaturnPos[:,1], '-',label='Saturn')
plt.plot(UranusPos[:,0], UranusPos[:,1], '-',label='Uranus')
plt.plot(NeptunePos[:,0], NeptunePos[:,1], '-',label='Neptune')
plt.plot(PerturberPos[:,0], PerturberPos[:,1], '-',label='Perturber')
plt.ylabel('y Position (AU)')
plt.xlabel('x Position (AU)')
plt.ylim( (-200, 200) )
plt.xlim( (-100, 300) )



plt.legend(loc=0)
plt.show()

EarthDis = np.sqrt(np.sum(EarthPos**2,axis=1))
MarsDis = np.sqrt(np.sum(MarsPos**2,axis=1))
JupiterDis = np.sqrt(np.sum(JupiterPos**2,axis=1))
SaturnDis = np.sqrt(np.sum(SaturnPos**2,axis=1))
UranusDis = np.sqrt(np.sum(UranusPos**2,axis=1))
NeptuneDis = np.sqrt(np.sum(NeptunePos**2,axis=1))
PerturberDis = np.sqrt(np.sum(PerturberPos**2,axis=1))

plt.plot(time, EarthDis, '-')#,label='Earth')
plt.plot(time, MarsDis, '-')#,label='Mars')
plt.plot(time, JupiterDis, '-',label='Jupiter')
plt.plot(time, SaturnDis, '-',label='Saturn')
plt.plot(time, UranusDis, '-',label='Uranus')
plt.plot(time, NeptuneDis, '-',label='Neptune')
plt.plot(time, PerturberDis, '-',label='Perturber')
plt.ylabel('Distance (AU)')
plt.ylim( (0, 250) )
plt.xlabel('Time (Yr)')
plt.legend(loc=0)
plt.show()


#Plotting Energy loss with time
EnergyData = open('../Data/EnergyStable.txt', 'r') 
Energies = []
for line in EnergyData.readlines():
    values = line.split()
    for i in range(len(values)):
        values[i] = float(values[i])
    Energies.append(values)
EnergyData.close()
Energies = np.array(Energies)

#'''
EnergyDataF = open('../Data/EnergyStableF.txt', 'r') 
EnergiesF = []
for line in EnergyDataF.readlines():
    values = line.split()
    for i in range(len(values)):
        values[i] = float(values[i])
    EnergiesF.append(values)
EnergyDataF.close()
EnergiesF = np.array(EnergiesF)
#'''


TotalEnergyLosses = Energies[:,1]
TotalEnergyLossesF = EnergiesF[:,1]
timeF = EnergiesF[:,0]
time = Energies[:,0]
plt.plot(time, TotalEnergyLosses, '-')#,label='Second Derivative')
#plt.plot(timeF, TotalEnergyLossesF, '-',label='First Derivative')
#plt.plot(time, KineticEnergyLosses, '-',label='Kinetic')
#plt.plot(time, PotentialEnergyLosses, '-',label='Potential')
plt.ylabel('Fractional Energy Loss')
plt.xlabel('Time (Yr)')
#plt.xlim( (400, 425) ) 
#plt.legend()
plt.show()


#Getting eccentricities with time and plotting
EccData = open('../Data/Eccentricities.txt', 'r') 
Eccs = []
for line in EccData.readlines():
    values = line.split()
    for i in range(len(values)):
        values[i] = float(values[i])
    Eccs.append(values)
EccData.close()
Eccs = np.array(Eccs)
time = np.array(Eccs[:,0])

EarthEcc = np.array(Eccs[:,1])
MarsEcc = np.array(Eccs[:,2])
JupiterEcc = np.array(Eccs[:,3])
SaturnEcc = np.array(Eccs[:,4])
UranusEcc = np.array(Eccs[:,5])
NeptuneEcc = np.array(Eccs[:,6])

#plt.plot(time, EarthEcc, '-',label='Earth')
#plt.plot(time, MarsEcc, '-',label='Mars')

#plt.plot(time, JupiterEcc, '-',label='Jupiter')
plt.plot(time, SaturnEcc, '-',label='Saturn')
plt.plot(time, UranusEcc, '-',label='Uranus')
plt.plot(time, NeptuneEcc, '-',label='Neptune')
plt.ylabel('Eccentricity')
plt.xlabel('Time (Yr)')
plt.xlim( 100 )
plt.legend()
plt.show()

#Getting eccentricity change with angle
AngleData = open('../F90 Code/OutputData/TotalsM5Vi10d180Rand.txt', 'r') 
Angles = []
for line in AngleData.readlines():
    values = line.split()
    for i in range(len(values)):
        values[i] = float(values[i])
    Angles.append(values)
AngleData.close()
Angles = np.array(Angles)
print(Angles.shape)

Degrees = Angles[19::20,0]
Saturn = Angles[19::20,1]
Uranus  = Angles[19::20,2]
Neptune = Angles[19::20,3]

plt.plot(Degrees, Saturn, '-',label='Saturn')
plt.plot(Degrees, Uranus, '-',label='Uranus')
plt.plot(Degrees, Neptune, '-',label='Neptune')
plt.ylabel('Max Eccentricity')
plt.xlabel('Angle of incidence (Degrees)')
plt.legend(loc=0)
plt.show()
```
